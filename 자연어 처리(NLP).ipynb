한국어의 토큰화 어려운 이유?

한국어는 교착어이다. - 실질적 의미를 가진 단어 또는 어간에 문법적인 기능을 가진 요소가 결합된 형태

그리고 한국어는 띄어쓰기가 잘 지켜지지 않는다 - 띄어쓰기를 잘 하지 않아도 이해가 된다

주어생략이 가능하고 어순이 중요하지 않다 - 그래도 이해가 된다

한자어라는 특성상 하나의 음절도 다른 의미를 가질 수 있다

교착어 - 실질적 의미를 가지는 어간에 조사나 어미와 같은 문법 형태소가 결합하여 문법적인 기능(각각이 다른 의미를 갖는)이 부여되는 언어

정제
- 불필요한 데이터 제거
- 특수기호 제거
- 불용어 제거
  - 불용어(stopword)? 자주등장하지만 분석을 하는 것에 있어서 큰 도움이 되지 않은 단어들
  - ex) I, my, me, over
  - NLTK에 불용어를 패키지 내에서 미리 정의하고 있음
  - 개발자가 임의로 불용어 지정 가능
- 텍스트의 인코딩 문제 해결
    - 인코딩과 디코딩?
    - 문자코드 : 컴퓨터는 문자를 인식할 수 없기 때문에 숫자로 변환되어 저장. 변환해주기 위한 기준이 필요. 그게 바로 문자코드.
    - 문자코드의 종류 : ASCII코드, 유니코드
    - 문자인코딩 : 문자 -> 코드 변환
    - 문자디코딩 : 코드 -> 문자 변환
- 텍스트를 코드로 변환하는 데 생길 수 있는 문제. 인코딩 문제를 미리 해결해줌
- 길이가 짧은 단어들은 의미를 가지지 않는 단어인 경우가 많기 때문에 길이가 짧은 단어 제거
- 등장 빈도가 적은 단어일수록 중요도가 떨어지는 단어이기 때문에 등장 빈도가 적은 단어 제거

정규화
- 문장의 복잡도를 줄여주는 과정
- 같은 의미를 가지고 있는 단어들을 1개로 통합하는 작업
- ex) am, are, is -> be
- ex) ㅋ, ㅋㅋㅋㅋ, ㅋㅋㅋㅋㅋㅋ -> ㅋ
- 대소문자 통합
- 이모티콘 -> emo

영어 문장 토큰화 실습
1. 기본 토크나이저
```python
import nltk
nltk.download("punkt")
from nltk.tokenize import word_tokenize
print(word_tokenize(sentence))
```

2. WordPunkTokenizer
- 어퍼스트로피 때문에 일어나는 일 해결
```python
from mltk.tokenize import WordPunctTokenizer
print(WordPunctTokenizer().tokenize(sentence))
```

3. TreebankWordTokenizer
```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(sentence))
```

````
# 토큰화 실습
# 영어 문장 토큰화 실습
import nltk
nltk.download("punkt")
# 영어 토크나이저 중 punkt 를 사용

# 토크나이저 여러가지 종류 있음
````

````
sentence = "Ain't nothin' sweeter, you want this sugar, don't ya?"
````

````
# 1. 기본 토크나이저
from nltk.tokenize import word_tokenize
print(word_tokenize(sentence))
````

````
# 2. WordPunkTokenizer
# 주로 ' 때문에 일어나는 일들을 해결해준다.
from nltk.tokenize import WordPunctTokenizer
print(WordPunctTokenizer().tokenize(sentence))
````

````
# 기본 토크나이저와의 차이점을 보면 기본 토크나이저와는 다르게
# 일반 띄어쓰기로 자르는 것 뿐만 아니라
# 어퍼스트로피의 앞과 뒤로도 자른다.
````

````
# 3. TreebankWordTokenizer
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(sentence))
````
