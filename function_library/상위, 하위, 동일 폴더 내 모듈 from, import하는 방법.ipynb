# 하위 폴더 내 파일, 동일 폴더 내 파일 참조

````
# 기준 : ee.py

# 참조 : aa.py
from sub1 import aa

# 참조 : ff.py
import ff
# 또는
from . import ff       ## from . 의미 : 동일폴더
````

# 상위 폴더 내 파일 참조

부모폴더의 절대경로를 참조 path에 추가

````
# 기준 : aa.py
# 참조 : cc.py
import os
import sys
sys.path.append(os.path.dirname(os.path.abspath(os.path.dirname(__file__))))

import cc.py
````

````
# 기준 : aa.py
# 참조 : gg.py
import os
import sys
sys.path.append(os.path.dirname(os.path.abspath(os.path.dirname(os.path.abspath(os.path.dirname(__file__))))))

import gg.py
````
