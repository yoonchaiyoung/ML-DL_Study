일반 텍스트 - 파일 입출력 활용

리스트나 클래스 같이 텍스트가 아닌 자료형 - pickle 모듈 사용

## pickle 모듈
- 원하는 데이터를 자료형의 변경X 파일로 저장하여 그대로 로드 가능
- 저장/로드할 때 파일을 바이트형식으로 읽고/쓰기
  - wb, rb
  - wb로 데이터 입력하는 경우 : .bin 확장자 사용하는 것이 좋음
- 모든 파이썬 데이터 객체를 저장/로드 가능

```
import pickle
```

```
# 일반적인 방법으로 데이터 입력
text = "hello world"
with open("hello.txt", "w") as f:
  f.write(text)

!cat hello.txt
```

```
# 입력 : pickle.dump(data, file)
list = ["a", "b", "c"]
with open("list.txt", "wb") as f:
  pickle.dump(list, f)
```

```
# 로드 : 변수명 = pickle.load(file)
# 한줄씩 파일 읽어옴
# 더이상 로드할 데이터 없으면 EOFError 발생
with open("list.txt", "rb") as f:
  data = pickle.load(f)
data
```
