# 시각화 분할
- subplot
  - 일일이 다 설정해줘야 한다.
```python
plt.subplot(2, 1, 1)  # 2행 1열의 그림 중 1번째 꺼
plt.subplot(2, 1, 2)  # 2행 1열의 그림 중 2번재 꺼
plt.show()
```

- subplots
  - 한 번에 설정할 수 있다.
```python
  fig, axes = plt.subplots(nrows=2, ncols=1)
  # fig : 전체 그림 통째로
  # axes : 그림 하나하나
  # nrows : 행 몇개
  # ncols : 열 몇개
  # sharex : xtick 값을 공유할지 안할지(x축에 표시하는 숫자들)
  # sharey : ytick 값을 공유할지 안할지(y축에 표시하는 숫자들)
  plt.show()
```
